## Data prep
- data used from kaggle by Bergur Tamimi

```python
import pandas as pd
```

```python
df = pd.read_csv("premier_league_odds.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
filtered_df = df[df['Date'] > '2022-01-01']
columns_to_drop = ['HomeTeamGoals', 'HomeTeamPoints', 'AwayTeamPoints', 'AwayTeamGoals', 'Round', 'HomeTeamScoredGoals', 'AwayTeamScoredGoals', 'HomeTeamConcededGoals', 'AwayTeamConcededGoals']
filtered_df = filtered_df.drop(columns=columns_to_drop)
filtered_df.to_csv('filtered_data.csv', index=False)
filtered_df.head()
```
